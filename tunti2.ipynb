```python
import matplotlib.pyplot as plt

def plot_data(df):
    plt.figure(figsize=(20, 10))
    for i, col in enumerate(df.select_dtypes(include=['int', 'float'])):
        plt.subplot(3,3, i+1)
        df[col].plot(kind='hist', title=col)
        plt.xlabel(col)
```

```python
plot_data(df)
```
